# Remove other format of pictures year

1. Remove the SVG format
2. Reomve the pictures that are not numeric

```python
import pandas as pd

# 读取第一个CSV文件
df1 = pd.read_csv('./art_label_7468.csv')

# 读取第二个CSV文件
df2 = pd.read_csv('./final_arts_info.csv')

# 使用pd.concat()函数按行对应将列拼接在一起
concatenated = pd.concat([df1, df2], axis=1)

# 打印拼接后的结果
concatenated
```

```python
filtered_concatenated = concatenated[~concatenated['link'].str.endswith('.svg')]

# 打印筛选后的结果
# print(filtered_concatenated)
filtered_concatenated
```

```python
filtered_concatenated['is_number'] = filtered_concatenated['year'].apply(lambda x: x.isdigit())
filtered_concatenated
```

```python
filtered_concatenated = filtered_concatenated[filtered_concatenated['is_number'] == True]
filtered_concatenated
```

```python
final = filtered_concatenated[['filename','year']]
final.to_csv('final_art_label.csv',index=False)
```
